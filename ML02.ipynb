```python
xi = [453, 11, 86]
```

```python
w0 = 0
w = [0.01, 0.04, 0.002]
```

```python
def linear_regression(xi):
    n = len(xi)

    pred = w0

    for j in range(n):
        pred = pred + w[j] * xi[j]
    
    return pred
```

```python
linear_regression(xi)
```
